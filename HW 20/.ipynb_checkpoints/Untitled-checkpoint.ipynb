```python
# Clustering/Dimensionality Reduction
from sklearn.manifold import TSNE
from sklearn.cluster import DBSCAN
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

from sklearn.datasets import load_wine

# DataScience
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('myopia.csv')
```

```python
df.info()
```

```python
# Check for obvious clusters
X = TSNE(learning_rate="auto").fit_transform(df)

plt.figure(figsize=(12,8))

plt.scatter(X[:,0], X[:,1])
plt.title("TSNE Plot")
plt.show()
```

```python
# Initialize PCA model
pca = PCA()
pca.fit(df)

plt.plot(range(len(df.columns)), pca.explained_variance_ratio_)
plt.xlabel("# of Principal Components")
plt.ylabel("Explained Variance")
plt.title("PCA Elbow")
plt.show()
```

```python
pca = PCA(n_components=3)
pca.fit(df)

components = pd.DataFrame(pca.transform(df), columns=["PCA1", "PCA2", "PCA3"])

components.head()
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(df)
preds = kmeans.predict(df)

plt.figure(figsize=(12,8))

plt.scatter(X[:,0], X[:,1], c=preds)

plt.title("TSNE Plot")
plt.show()
```

```python
inertia = []
k_range = list(range(2, 10))

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(components)
    
    inertia.append(kmeans.inertia_)
    
# make plot
plt.figure(figsize=(10,6))

plt.plot(k_range, inertia)
plt.xlabel("Number of Clusters")
plt.ylabel("Inertia")
plt.title("Elbow Plot")

plt.show()
```

```python
kmeans = KMeans(n_clusters=7, random_state=42)
kmeans.fit(df)
preds = kmeans.predict(df)

plt.figure(figsize=(12,8))

plt.scatter(X[:,0], X[:,1], c=preds)

plt.title("TSNE Plot")
plt.show()
```

```python
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans.fit(df)
preds = kmeans.predict(df)

plt.figure(figsize=(12,8))

plt.scatter(X[:,0], X[:,1], c=preds)

plt.title("TSNE Plot")
plt.show()
```

```python
mergings = linkage(components, method="ward")

plt.figure(figsize=(20,8))

dendrogram(mergings,
          leaf_rotation=90,
          leaf_font_size=5)

plt.show()
```

```python
cluster = AgglomerativeClustering(
    n_clusters=5,
    affinity="euclidean",
    linkage="ward"
)

preds = cluster.fit_predict(df)
plt.figure(figsize=(12,8))

plt.scatter(X[:,0], X[:,1], c=preds)
plt.title("TSNE Plot")
plt.show()
```

Either 4 or 5 clusters seems to be the safe. The elbow plot is more of a curve then a singular point. It seems that once you go past 5 and especially 6, the TSNE plots become confluded and hard to read. As my knowledge base is limited I would say now that somewhere in the middle seemms like a reasonable amount.
