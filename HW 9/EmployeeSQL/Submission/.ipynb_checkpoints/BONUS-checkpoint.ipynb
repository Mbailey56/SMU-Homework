```python
inspector_gadget = inspect(engine)

tables = inspector_gadget.get_table_names()
for table in tables:
    print(table)
    
    # get all columns in table
    columns = inspector_gadget.get_columns(table)
    for column in columns:
        print(column)
    print()
```

```python
query = """SELECT
                d.dept_no,
                d.dept_name,
                e.last_name,
                e.first_name,
                t.title,
                salary
            FROM
                departments d
                join dept_manager dm on d.dept_no = dm.dept_no
                join employees e on dm.emp_no = e.emp_no
                join titles t on e.emp_title_id = t.title_id
                join salaries s on s.emp_no = e.emp_no
            order by
                d.dept_no,
                e.last_name;"""

df = pd.read_sql(query, engine)
df.dropna()
df.head()
```

```python
salaries = df.groupby("dept_name").salary.mean().sort_values(ascending=False)
salaries
```

```python
fig,ax=plt.subplots(figsize=(14,5))

plt.barh(salaries.index,salaries.values, color = "indigo")
plt.title("Manager Salaries by Department")
ax.set_xlabel("Salaries")
ax.set_ylabel('Department')
plt.show()
```

```python
query_salary = """SELECT
                            e.emp_no as employee_number,
                            e.last_name,
                            e.first_name,
                            e.sex,
                            s.salary
                     FROM
                            employees e
                            join salaries s on e.emp_no = s.emp_no
                     ORDER BY
                            e.last_name asc;"""
df2 = pd.read_sql(query_salary, engine)
df2.dropna()
df2.head(20)
```

```python
df2.max()
```

```python
df2.sort_values(by=["salary"],ascending=True)
```

```python
df2["salary range bin"]=None
df2.loc[(df2.salary >= 40000) & (df2.salary <= 60000), "salary range bin"] = "40000-60000"
df2.loc[(df2.salary > 60000) & (df2.salary <= 80000), "salary range bin"] = "60000-80000"
df2.loc[(df2.salary > 80000) & (df2.salary <= 100000), "salary range bin"] = "80000-100000"
df2.loc[(df2.salary > 100000) & (df2.salary <= 120000), "salary range bin"] = "100000-120000"
df2.loc[(df2.salary > 120000) & (df2.salary <= 140000), "salary range bin"] = "120000-140000"

df2.head()
                
```

```python
plt.figure(figsize=(14,6))
sns.histplot(data=df2, x="salary range bin")
plt.title("Salary Ranges")
plt.show()
#40000-60000 dollars salary range is the highest
```
